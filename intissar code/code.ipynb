```python
import math
D1= input("donner le diamétre de section 1 en m")
V1=input("donner le vitesse de la section 1 en m/s")
MV=input("donner le masse volumique de l'eau qui circule ")
D1=float(D1)
V1=float(V1)
MV=float(MV)
V2=1.5*V1
alpha=10
B=math.tan(alpha)
C=math.sqrt(V1/V2)
L=((-D1)/2*B)*(C-1)
dp= 0.5*MV*((V1*2)-(V2*2))
print("la longueur de la convergent pour que la vitesse multiplié avec 1.5 est :",L,"m")
print("la différence de la pression entre l'entrée et sortie (P2-P1) est :",dp,"Pa")
```
